```python
import pandas as pd
```

```python
data = pd.read_csv("closed_caption_12_30_21_non_CO.csv")
```

```python
print(data.head())
```

```python
filtered_data = data[data['data'] == 'yes']
```

```python
filtered_data.head(3)
```

```python
filtered_data['Closed Captions']
```

```python
filtered_data['captions'] = filtered_data['Closed Captions'].str.strip('>')
filtered_data['captions'] = filtered_data['captions'].str.strip()
```

```python
filtered_data['captions'].values
```

```python
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```python
words = " ".join(word for word in filtered_data['captions'])
stopwords = set(STOPWORDS)
```

```python
stopwords
```

```python
cap_stopwords = []
for word in stopwords:
    cap_stopwords.append(word.upper())
```

```python
cap_stopwords.sort()
```

```python
cap_stopwords = set(cap_stopwords)
```

```python
cap_stopwords
```

```python
words = words.replace('.', '').replace(',', '').replace('â€™',"'")
```

```python
words
```

```python
len(stopwords)
```

```python
len(cap_stopwords)
```

```python
len(words.split())
```

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='black', colormap='Set2', collocations=False, stopwords = cap_stopwords).generate(words)

# Plot
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
clean_words = []

for word in words.split():
    if word not in cap_stopwords:
        clean_words.append(word)
```

```python
clean_words
```

```python
def word_freq(words):
    count = {}

    for word in words:
        if word not in count:
            count[word] = 1
        else:
            count[word] += 1
            
    return count
```

```python
word_counts = word_freq(clean_words)
word_counts
```

```python
words_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Count'])

final_words_df = words_df[words_df['Count'] > 1]
final_words_df
```

```python
# set figure size
fig, ax = plt.subplots(figsize=(12, 8))
# plot horizontal bar plot
final_words_df.sort_values(by='Count').plot.barh(x="Word", y="Count", ax=ax)
# set the title
plt.title("Most Common words in captions")

for i, v in enumerate(final_words_df['Count'].sort_values()):
    ax.text(v + 1, i - .2, str(v),
            color = 'blue', fontweight = 'bold')

plt.show()
```

```python
clean_words_str = " ".join(word for word in final_words_df['Word'].values)
clean_words_str
```

```python
wordcloud = WordCloud(width = 3000, height = 2000, background_color='black', colormap='Set2', collocations=False, stopwords = cap_stopwords).generate(clean_words_str)

# Plot
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
